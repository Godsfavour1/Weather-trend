# DATA plots on Line Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize =[15,10])
plt.plot(cleaned_berlin_data['year'], cleaned_berlin_data['avg_temp'] )

plt.plot(global_data['year'], global_data['avg_temp'] )

plt.legend(['Berlin Data', 'Global_data'])
plt.xlabel('Year')
plt.ylabel('Avg Temp')
```

# Add moving average columns

```python
#To Berling Data
cleaned_berlin_data['adjusted_year'] = cleaned_berlin_data['year'].rolling(window = 8).mean()
cleaned_berlin_data['adjusted_avg_temp'] = cleaned_berlin_data['avg_temp'].rolling(window = 8).mean()


#To Global Data
global_data['adjusted_year'] = global_data['year'].rolling(window = 8).mean()
global_data['adjusted_avg_temp'] = global_data['avg_temp'].rolling(window = 8).mean()
```

```python
cleaned_berlin_data
```

# Line graph with moving averages

```python
import matplotlib.pyplot as plt

plt.figure(figsize =[15,10])
plt.plot(cleaned_berlin_data['adjusted_year'], cleaned_berlin_data['adjusted_avg_temp'] )

plt.plot(global_data['adjusted_year'], global_data['adjusted_avg_temp'] )

plt.legend(['Berlin Data', 'Global_data'])
plt.xlabel('Year')
plt.ylabel('Avg Temp')
```

# Observations from Line chart

1. The Berlin average temperates are relatively higher than the global average temperatures.
